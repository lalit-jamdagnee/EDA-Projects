```python
df_country = pd.read_excel('Dataset/Zomatodataset/Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on="Country Code", how="left")
```

```python
final_df.head()
```

```python
# to check data types
final_df.dtypes
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
# Plotting pie chart for top 3 countries that uses zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct = '%1.2f%%' )
```

Observation: Zomato's maximum records or transaction are from India, followed by United States and United Kingdom

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings
```

### Observation

1. When Rating is b/w 4.5 to 4.9 --> Excellent
2. when Rating are b/w 4.0 to 4.4 --> Very Good
3. when Rating is b/w 3.5 to 3.9 --> Good
4. when Rating is b/w 2.5 to 3.4 --> Average
5. when Rating is b/2 1.9 to 2.4 --> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating", y = "Rating Count", data=ratings)
```

```python
plot = sns.barplot(x = "Aggregate rating", y="Rating Count", data = ratings, hue="Rating color", palette=['white','red','orange','yellow','green','green'])
plt.show()
```

### Observation

1. Not rated count is very high
2. Maximum nuber of rating are between 2.5 to 3.4

```python
# Count plot
sns.countplot(x="Rating color", data = ratings,hue="Rating color", palette=['white','red','orange','yellow','green','green'])
```

Its not rating count, it's the frequency count of the rating colors
